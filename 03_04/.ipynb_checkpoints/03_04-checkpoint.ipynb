```python
# Step 1: Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Step 2: Load an image
image_path = 'palmtree.png'  # Replace with the path to your image
image = cv2.imread(image_path)
```

```python
# Step 3: Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Step 4: Apply Gaussian blur to the image
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
# Step 5: Detect edges using the Canny edge detection algorithm
edges = cv2.Canny(blurred_image, 50, 150)
```

```python
# Step 6: Display the original and processed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edge Detection')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
```
